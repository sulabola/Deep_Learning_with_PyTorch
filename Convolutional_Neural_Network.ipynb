#### Convolutional Nueural Network (CNN)

CNNs are great with classifiying images.

General steps:
Image input --> Featured maps --> Pooled Featured maps --> ... --> Flatten layer -->Fully connected layer --> Output

Link: https://www.tech-teacher.jp/blog/wp-content/uploads/2023/09/2.png

For this we will be using a mnist dataset (that is, database of handwritten digits). Link: https://www.tensorflow.org/datasets/catalog/mnist

We may be lokking at a hand written number written in color white on a black paper. When convert it into numbers (a matrix), the black may take number 0 and white may take number 1. On edges, where it is shaddy, it may take a number between 0 and 1 (greater than 0) / shade of gray. A computer identifies the differences (feature mapping) and the neural network would be able to classify the image (identify the number). Note: This is a classification problem and we will train the model with train data set which contains labels.

#### Image Filter / Image Kernal Overview

We can view an image as a matrix of ones and zeros. Let's say we want to blur a part of an image. Then what we do is, take that part of images/matrix called image patch (local receptive field) and use image filter / image kernal (we have different kind of filters). Then we end up with a new matrix and we will multiple that with the image patch. Then take the sum of the values and replace the original block (so we may see some dimention reduction. If we start with 6X6 matrix, with 3X3 blocks, we end up with 4X4 matrix). Then, we will apply this pixcel by pixcel / block by block. The term for dragging that cross the image is called convolution. The neural network will learn the best numbers for the kernal over the time. Link: https://iq.opengenus.org/content/images/2022/12/cnn.png

Interesting web link to see how kernals work: https://setosa.io/ev/image-kernels/

#### Convolutional Layer and RGB

One of the important question is why we do use CNN instead of ANN.

Note that for a big image (even colored image), it has much more details and we end up with significant number of outputs (consider end up with a 4X4 matricx starting with 6X6 matrx). In a fully connected ANN, this requires significantly high computer power. In CNN, the convolution layer is not fully connected (it is locally connected).

For color images, the filtering process is slightly different. Becase color images are 3-D images (tensors). Color images are made with 3 color channels (RGB). Intesity determine the color. For basic images, in CNN we have height width and color channel. For color images, we have 3 different color channels. Then we will have: height+width+R, height+width+G, height+width+B. Thus, each color channel become another filter layer.

#### Pooling Layer in Convolutional Neural Network

Pooling layer: It takes all the inputs from convolution layer and pools them and reuduce them. Theoratically we can do that with regular ANN. However, there are so many paramters/weights

Convolutional layer: Original Image --> Filter --> Convolutional Layer (That is, 6X6 image to 4X4 image)

Pooling layer: Reduce convolution layer further (down sampling/sub sampling) (that is, 4X4 --> 2X2).

There are several methods to down sample/sub sample. Popular methods include: max pooling, average pooling. When we do this we are loosing some information. Link: https://iq.opengenus.org/content/images/2022/12/download.png

```python
### Define CNN Model

### Describe convolutional layer and what is it doing (2 convolutional layers)

### First convolutional layer
conv1 = nn.Conv2d(1, 6, 3, 1)

### 4 important paramters: input size (here 1 image), output size (Here we picked 6. It could be somthing else), kernal size (3 or 3X3 matrix), stride (1)
### stride = 1 means we move one at a time. That is we move/shift one column at a time.

### Second convolutional layer
conv2 = nn.Conv2d(6, 16, 3, 1)
### Here number of inputs is 6. This is because in the previous layer we have 6 outputs.
### output size is 16 (random choice)
```
