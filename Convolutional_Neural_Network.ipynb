#### Convolutional Nueural Network (CNN)

CNNs are great with classifiying images.

General steps:
Image input --> Featured maps --> Pooled Featured maps --> ... --> Flatten layer -->Fully connected layer --> Output

For this we will be using a mnist dataset (that is, database of handwritten digits). Link: https://www.tensorflow.org/datasets/catalog/mnist

We may be lokking at a hand written number written in color white on a black paper. When convert it into numbers (a matrix), the black may take number 0 and white may take number 1. On edges, where it is shaddy, it may take a number between 0 and 1 (greater than 0) / shade of gray. A computer identifies the differences (feature mapping) and the neural network would be able to classify the image (identify the number). Note: This is a classification problem and we will train the model with train data set which contains labels.

#### Image Filter / Image Kernal Overview

We can view an image as a matrix of ones and zeros. Let's say we want to blur a part of an image. Then what we do is, take that part of images/matrix called image patch (local receptive field) and use image filter / image kernal (we have different kind of filters). Then we end up with a new matrix and we will multiple that with the image patch. Then take the sum of the values and replace the original block (so we may see some dimention reduction. If we start with 6*6 matrix, with 3*3 blocks, we end up with 4*4 matrix). Then, we will apply this pixcel by pixcel / block by block. The term for dragging that cross the image is called convolution. The neural network will learn the best numbers for the kernal over the time.
