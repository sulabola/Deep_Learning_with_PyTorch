#### Tensor Operations

```python
my_torch = torch.arange(10)
my_torch
```

Reshape and View

```python
my_torch = my_torch.reshape(2,5)
my_torch
```

Reshape if we do not know dimentions exactly

```python
my_torch2 = torch.arange(12)
my_torch2
```

```python
my_torch2 = my_torch2.reshape(3,-1)
my_torch2
```

```python
my_torch2 = torch.arange(12)
my_torch2 = my_torch2.reshape(-1,3)
my_torch2
```

```python
my_torch3 = torch.arange(10)
my_torch4 = my_torch3.view(2,5)
my_torch4

### Note: "view" works same as "reshape"
### Differences = Link: https://www.geeksforgeeks.org/deep-learning/whats-the-difference-between-reshape-and-view-in-pytorch/
```

```python
### With reshape and view they will update

my_torch5 = torch.arange(10)
my_torch5
```

```python
my_torch6 = my_torch5.reshape(2,5)
my_torch6
```

```python
my_torch5[1] = 4141 ### Update the second element
my_torch5
```

```python
my_torch6

### Note: It has updated
```

Slices

```python
my_torch7 = torch.arange(10)
my_torch7
```

```python
my_torch7[7]

### Note: It is not grabing the number. It is a tensor at index 7
```

```python
my_torch8 = my_torch7.reshape(5,2)
my_torch8
```

```python
my_torch8[:,1] ### Allthe rows of column 1.

### Note: The output looks like a row vector.
```

```python
### Return a column (like a column vector)

my_torch8[:,1:]
```

#### Tensor Math Operations

```python
tensor_a = torch.tensor([1,2,3,4])
tensor_b = torch.tensor([5,6,7,8])
```

```python
### Addition (element by element)

tensor_a + tensor_b
```

```python
### Addition (longhand)

torch.add(tensor_a, tensor_b)
```

```python
### Substraction

tensor_b - tensor_a
```

```python
### Substraction (function)

torch.sub(tensor_b, tensor_a) ### "torch.substract" is an aliace
```

```python
### Multiplication

tensor_a * tensor_b
```

```python
torch.mul(tensor_a, tensor_b)
```

```python
### Division

tensor_a / tensor_b
```

```python
torch.div(tensor_a, tensor_b)
```

```python
### Modulus (remainder after division)

tensor_b % tensor_a
```

```python
torch.remainder(tensor_b, tensor_a)
```

```python
### Exponent / Power

torch.pow(tensor_a, tensor_b)
```
